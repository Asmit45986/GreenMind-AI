```python
import os
import sys

# Project root linking
NOTEBOOK_DIR = os.getcwd()
PROJECT_ROOT = os.path.abspath(os.path.join(NOTEBOOK_DIR, ".."))
if PROJECT_ROOT not in sys.path:
    sys.path.insert(0, PROJECT_ROOT)

from models.recommendation.survival_engine import TreeSurvivalPredictor

# 1. Predictor engine instance create karein
predictor = TreeSurvivalPredictor()

# 2. Dynamic Environment data check (Same profile context)
mock_env_profile = {
    "weather": {"temperature": 32.5, "humidity": 45.0},
    "air_quality": {"aqi_index": 3}
}

# 3. Test list for unique recommendation trees
test_trees = ["Neem", "Acacia (Babool)", "Shisham", "Bamboo", "Khejri", "Amaltas"]

print("="*65)
print("📈 AI TREE SURVIVAL PROBABILITY REPORT")
print("="*65)
print(f"🌡️ Temperature: {mock_env_profile['weather']['temperature']}°C | 💧 Humidity: {mock_env_profile['weather']['humidity']}% | 💨 AQI: {mock_env_profile['air_quality']['aqi_index']}")
print("-"*65)

for tree in test_trees:
    res = predictor.predict_survival(tree_name=tree, env_profile=mock_env_profile)
    print(f"🌳 {tree:15} -> Survival Chance: {res['survival_percentage']}% | Confidence: {res['confidence']}%")
    print(f"   Status: {res['risk_level']}\n")
    
print("="*65)
```
