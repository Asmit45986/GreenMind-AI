```python
import os
import sys

# Project root validation
NOTEBOOK_DIR = os.getcwd()
PROJECT_ROOT = os.path.abspath(os.path.join(NOTEBOOK_DIR, ".."))
if PROJECT_ROOT not in sys.path:
    sys.path.insert(0, PROJECT_ROOT)

from models.recommendation.carbon_engine import CarbonCreditEstimator

# 1. Initialize Carbon Engine
estimator = CarbonCreditEstimator()

# 2. Mock Pipeline inputs (Supposing user plants 500 trees per category)
tree_plantation_count = 500
survival_data = {
    "Neem": 95.0,
    "Acacia (Babool)": 87.5,
    "Shisham": 95.0,
    "Bamboo": 85.25,
    "Khejri": 77.75,
    "Amaltas": 80.0
}

print("="*75)
print(f"💰 AI CARBON CREDIT & REVENUE PROJECTION (10 YEAR HORIZON)")
print(f"🌲 Planned Plantation: {tree_plantation_count} Trees per species")
print("="*75)

for tree, survival in survival_data.items():
    metrics = estimator.estimate_carbon_credits(
        tree_name=tree, 
        tree_count=tree_plantation_count, 
        survival_percentage=survival
    )
    
    print(f"🌳 {tree:15} | Active Trees: {metrics['effective_active_trees']}/{tree_plantation_count}")
    print(f"   💨 CO₂ Sequestered : {metrics['total_co2_absorbed_tons']} Tons")
    print(f"   🪙 Carbon Credits  : {metrics['carbon_credits']} Credits")
    print(f"   💸 Est. Return (₹) : ₹{metrics['estimated_revenue_inr']:,}\n")

print("="*75)
```
